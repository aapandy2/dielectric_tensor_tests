```python
x = np.linspace(0., 100000., 1000)

a = 0.000001
b = 10.

def f(x):
    ans = np.exp(-a*x)
    return ans

pl.plot(x, f(x) * np.sin(b*x))
quad_result = quad(f, 0., np.inf, weight='sin', wvar=b)

print 'quad:    ', quad_result[0], 'error: ', quad_result[1]
print 'analytic: ' + str(b / (a**2. + b**2.))
print 'limit:   ', 1./b
```
